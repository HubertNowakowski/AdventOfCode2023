```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

$t$ - total race time \
$p$ - time pressing the button (charging the boat) \
$v$ - speed of the boat

pressing the button for time $p$, sets the speed to: 
> $v=p$

but the remaining time is 
> $t_{1} = t-p$

the distance traveled d is given by:
> $d = v * t_{1} = p*(t-p)$

```python
t = 46_689_866
max_d = 358105418071080

step = 1e3

p = np.arange(0,t+1, step)
d = p*(t-p)

f, ax = plt.subplots()
ax.plot(p, d, '.', markersize=1)
ax.axhline(y=max_d, color='r', linestyle='--')
plt.xlabel('Time holding the button p [ms]')
plt.ylabel('Distance d [mm]')
plt.show()

print(f"Points above the record: {sum(d>max_d)}")
```

solving the quadratic equation, because theres too many points to compare every each one:

>$ d_{max} = p \cdot (t-p) $ \
$ p^2 - t \cdot p + d_{max} = 0$ \
$ \Delta = t^2 - 4 \cdot 1 \cdot d_{max} $ \
\
$ p_1 = \frac{t - \sqrt{t^2 - 4d_{max}}}{2} $ 
$ p_2 = \frac{t + \sqrt{t^2 - 4d_{max}}}{2} $ 


```python
p1 = t/2 - math.sqrt( t**2 - 4*max_d )/2
p2 = t/2 + math.sqrt( t**2 - 4*max_d )/2
print( p1, p2)
print( np.ceil(p1), np.floor(p2) )
print( int( np.floor(p2) - np.ceil(p1) +1))
```
